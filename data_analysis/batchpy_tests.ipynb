# Teste das rotinas da solução analítica

```python
from batchpy import batch as bh
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
def animate_particle(a, bi, eps, s, outputfile):
    plt.style.use('seaborn-pastel')
    lines = ()
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
    ax.set(xlim=(0.0, 1.0), ylim=(0.0, 1.0))
    ax.set(xlabel='r/R', ylabel='u')

    lines = ()
    for i in range(len(bi)):
        for j in range(len(a)):
            line, = ax.plot([],[])
            lines = lines + (line,)

    def animate(frame, a, bi, s, eps):
        x = np.linspace(0, 1, 100)
        for i in range(len(bi)):
            for j in range(len(a)):
                y = bh.batch_particle(x,np.array([frame/100]),bi[i],a[j],s,eps)
                lines[i+j].set_data(x, y)
        return lines
   

    anim = FuncAnimation(fig, animate, fargs=(a, bi, s, eps), frames=100, interval=150, blit=True)

    anim.save(outputfile)
```

```python
a = [1]
bi = [0.1, 1, 10, 100]
s = 2
eps = 1e-6
outputfile = 'output.gif'
animate_particle(a, bi, eps, s, outputfile)
```
