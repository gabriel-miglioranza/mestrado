```python
from batchpy import batch as bh
import numpy as np
from matplotlib import pyplot as plt
```

```python
print(bh.__doc__)
```

```python
t = np.linspace(0.0,1,100)
#t = t**2
eps=1e-7

```

```python
bi = 1
a = 1
s = 2


def plot_bulk(t,bi,a,s):
    v=np.zeros(t.shape)
    eps = 1e-6
    for i in range(len(t)):
        v[i] = bh.finite_batch_bulk(t[i],bi,a,s,eps)
    plt.plot(t,v)
```

```python
plot_bulk(t,bi,a,s)
plot_bulk(t,100,a,s)
plot_bulk(t,1000,a,s)
plot_bulk(t,-1,a,s)
plot_bulk(t,-1,0,s)
plt.show()
```

```python
plot_bulk(t,bi,a,s)
plot_bulk(t,bi,2,s)
plot_bulk(t,bi,5,s)
plot_bulk(t,bi,0.2,s)
plot_bulk(t,bi,0,s)
plot_bulk(t,-1,0,s)
plt.show()
```
